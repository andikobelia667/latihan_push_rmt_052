```python
#Given list of Data:

a = [1, 4, 9, 16, 25, 36, 49,  64, 81, 100]

#print 16

print (a[3])

#print [36, 49, 64, 81]

print(a[5:9])

#kenapa 9 bukan 8, soalnya yang belakang itu stopnya dimana kan harusnya variabel 9 itu 100
#jadi stop di 100 dan 100nya gaikut masuk

#print [100, 81, 64, 49, 36, 25, 16, 9, 4, 1]
print(a[::-1])


print('Coba lain')
# Ambil dari index 0 sampai 3 (tidak termasuk 3)
print(a[0:3])      # [1, 4, 9]

# Ambil dari index 2 sampai akhir
print(a[2:])       # [9, 16, 25, 36, 49, 64, 81, 100]

# Ambil dari awal sampai index 5 (tidak termasuk 5)
print(a[:5])       # [1, 4, 9, 16, 25]

# Ambil semua dengan step 2 (loncat 1)
print(a[::2])      # [1, 9, 25, 49, 81]

# Ambil 3 elemen terakhir
print(a[-3:])      # [64, 81, 100]

# Ambil semua kecuali 2 elemen terakhir
print(a[:-2])      # [1, 4, 9, 16, 25, 36, 49, 64]

# Terbalik
print(a[::-1])     # [100, 81, 64, 49, 36, 25, 16, 9, 4, 1]

# Ambil elemen genap (index 1, 3, 5, ...)
print(a[1::2])     # [4, 16, 36, 64, 100]
```
